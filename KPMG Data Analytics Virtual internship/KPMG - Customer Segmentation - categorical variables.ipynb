# Building model

### Using K-means clustering

```python
final_df
```

```python
from sklearn.cluster import KMeans
X= final_df[['past_3_years_bike_related_purchases','Age','wealth_segment_new','property_valuation','tenure_new',
               'Manufacturing','Financial Services','Health','Retail','Property','IT','Entertainment','Argiculture',
                'Telecommunications']]
```

```python
X
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(X)
    sse.append(km.inertia_)
```

```python
plt.xlabel('k')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

```python
from sklearn.model_selection import cross_validate, cross_val_score, RepeatedKFold, StratifiedKFold
cv = RepeatedKFold(
        n_splits=5,
        n_repeats=10,
        random_state=420)
kmeans = KMeans(n_clusters=8, init='k-means++', max_iter=10, n_init=10, random_state=420)
cluster=cross_validate(kmeans, X, cv=cv, n_jobs=2)
```

```python

# Fit and predict 
y_means = kmeans.fit_predict(X)
```

```python
y_means
```

```python
plt.scatter(X["wealth_segment_new"], 
            X["tenure_new"], 
            c=y_means, s=20, cmap='viridis')
plt.xlabel("wealth_segment_new")
plt.ylabel("tenure_new")
plt.title("Customers K-MEANS Clustering");
```

```python
centroids = pd.DataFrame(kmeans.cluster_centers_, columns = ['past_3_years_bike_related_purchases','Age','wealth_segment_new','property_valuation','tenure_new',
               'Manufacturing','Financial Services','Health','Retail','Property','IT','Entertainment','Argiculture',
                'Telecommunications'])
```

```python
centroids.index_name = "ClusterID"
centroids["ClusterID"] = centroids.index
centroids = centroids.reset_index(drop=True)
```

```python
centroids
```

```python
X['label'] = y_means
X
```
